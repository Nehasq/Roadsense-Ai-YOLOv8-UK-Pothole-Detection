###  1) Using leave one out technique on 21 images (without augmentation)

Source: https://docs.ultralytics.com/guides/kfold-cross-validation/#k-fold-dataset-split



```python
 #Retrieving all label files for your dataset

dataset_path = Path("/content/drive/MyDrive/Colab Notebooks(NEHA)/Datasets/UK_dataset_new_pothole/valid")  # replace with 'path/to/dataset' for your custom data
labels = sorted(dataset_path.rglob("*labels/*.txt"))  # all data in 'labels'
```

```python
labels
```

```python
#Now, read the contents of the dataset YAML file and extract the indices of the class labels.

# Define the YAML file path
yaml_file = "/content/drive/MyDrive/Colab Notebooks(NEHA)/Datasets/UK_dataset_new_pothole/data_test_3.yaml"

# Open the YAML file and extract class labels
with open(yaml_file, "r", encoding="utf8") as y:
    data = yaml.safe_load(y)  # Use yaml.safe_load, not y.safe_load
    classes = data["names"]


print("Class Indices:", classes)
```

```python
#Initialize an empty pandas DataFrame
indx = [label.stem for label in labels]  # uses base filename as ID (no extension)
labels_df = pd.DataFrame([], columns=classes, index=indx)
```

```python
labels_df
```

```python
for label in labels:
    with open(label, "r") as lf:
        lines = lf.readlines()
    lbl_count = len(lines)  # Total number of objects in the file
    labels_df.loc[label.stem] = lbl_count

```

```python
labels_df
```

### K-Fold Dataset Split

Leave-one-out cross-validation, or LOOCV, is a configuration of k-fold cross-validation where k is set to the number of examples in the dataset.

```python
#Uisng the KFold class from sklearn.model_selection to generate k splits of the dataset.

ksplit = 21 # k -> equal to the no of images in the dataset
kf = KFold(n_splits=ksplit, shuffle=True, random_state=20)  # setting random_state for repeatable results

kfolds = list(kf.split(labels_df))
```

```python
kfolds
```

```python
#The dataset has now been split into k folds, each having a list of train and val indices. We will construct a DataFrame
#to display these results more clearly.

folds = [f"split_{n}" for n in range(1, ksplit + 1)]
folds_df = pd.DataFrame(index=indx, columns=folds)

for idx, (train, val) in enumerate(kfolds, start=1):
    folds_df[f"split_{idx}"].loc[labels_df.iloc[train].index] = "train"
    folds_df[f"split_{idx}"].loc[labels_df.iloc[val].index] = "test"
```

```python
folds_df
```

```python
#Next, we create the directories and dataset YAML files for each split.
import datetime

supported_extensions = [".jpg", ".jpeg", ".png"]

# Initialize an empty list to store image file paths
images = []

# Loop through supported extensions and gather image files
for ext in supported_extensions:
    images.extend(sorted((dataset_path / "images").rglob(f"*{ext}")))

# Create the necessary directories and dataset YAML files (unchanged)
save_path = Path(dataset_path / f"{datetime.date.today().isoformat()}_{ksplit}-Fold_Cross-val")
save_path.mkdir(parents=True, exist_ok=True)
ds_yamls = []

for split in folds_df.columns:
    # Create directories
    split_dir = save_path / split
    split_dir.mkdir(parents=True, exist_ok=True)
    (split_dir / "train" / "images").mkdir(parents=True, exist_ok=True)
    (split_dir / "train" / "labels").mkdir(parents=True, exist_ok=True)
    (split_dir / "test" / "images").mkdir(parents=True, exist_ok=True)
    (split_dir / "test" / "labels").mkdir(parents=True, exist_ok=True)

    # Create dataset YAML files
    dataset_yaml = split_dir / f"{split}_dataset.yaml"
    ds_yamls.append(dataset_yaml)


    with open(dataset_yaml, "w") as ds_y:
        yaml.safe_dump(
            {
                "path": split_dir.as_posix(),
                "train": "train",
                "val": "test",
                "names": classes,
            },
            ds_y,
        )
```

```python
ds_yamls
```

```python
#Lastly, copy images and labels into the respective directory ('train' or 'val') for each split.
import shutil

for image, label in zip(images, labels):
    for split, k_split in folds_df.loc[image.stem].items():
        # Destination directory
        img_to_path = save_path / split / k_split / "images"
        lbl_to_path = save_path / split / k_split / "labels"

        # Copy image and label files to new directory (SamefileError if file already exists)
        shutil.copy(image, img_to_path / image.name)
        shutil.copy(label, lbl_to_path / label.name)

```
